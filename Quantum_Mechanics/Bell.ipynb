# Desigualdad de Bell en funci√≥nde la temperatura.

```python
import numpy as np
import sympy as sym
from sympy.utilities.lambdify import lambdify
import matplotlib.pyplot as plt

sym.init_printing()  # LaTeX like pretty printing for IPython
```

```python
b,omega,t,A,Omega=sym.symbols(r'\beta \omega t A \Omega')
J=A*omega*sym.exp(-omega/Omega)
fun=J*(2*sym.sin(omega*t/2)/omega)**2*sym.coth(omega*b/2)
```

```python
f = lambdify([omega, A,Omega], J)
w=np.linspace(0,5,100)
Omega_=1
A_=1
tot=f(w,A_,Omega_)
plt.plot(w,tot,'k')
plt.xlabel('$\omega$',fontsize=20)
plt.ylabel('$J(\omega)$',fontsize=20)
```

```python
sym.integrate(fun,(omega,0,sym.oo))
```
