# Calculo de propiedades termodinamicas.

En física estadistica para un sistema en equilibrio, constituido por N particulas, existe un objeto que relaciona las cantidades microscopicas del sistema con la propiedades termodinamicas del mismo. Este objeto es conocido como la función partición, el cuál viene dado por:


$$Z = \sum_\epsilon  \text{exp}\left(-\beta \epsilon\right),$$

donde $\beta=1/k_bT$ y $\epsilon$ son las autoenergias del sistema. Nuestro sistema de estudio, es un gas ideal confinado en las dos siguientes geometrias:

# PEGAR IMAGEN DE LAS DOS GEOMETRIA

Al considerar que el sistema es no interactuante, la función partición puede ser calculada de forma más sencilla, esto es, ya que la energía total del sistema es separable en la contribución de energía cinetica de cada una de las particulas. Por tanto:

$$Z = \left \{ \sum_\epsilon  \text{exp}\left(-\beta \epsilon\right) \right\}^N /N!,$$

donde el factor $N!$ es añadido al considerar la indistinguibilidad de las particulas. En principio, ya que las particulas estan confinadas en 3D se debería de resolver la ecuación de Schrodinger para 3 dimensiones. Sin embargo, en el articulo, se usa una aproximación haciendo referencia a la hipotesis de Weyl, donde se aproxima las autoenergias en la dirección longitudinal, llegando a,

$$\zeta=\left(\frac{\sqrt{\pi}}{2 \alpha_l}-\frac{1}{2}\right)\sum_{k}\text{exp}\left(-\frac{\alpha_{t}^2}{\pi^2} \tilde{k}^2 \right),$$

y la función partición total $ Z= \zeta^{N}/N!$. Se definen los parametros $\alpha_t = L_c/L_*$, $\alpha_l=L_c/L_l$ y $\tilde{k}= \frac{2m\epsilon_{2D}}{\hbar^2}$. Aqui  $L_c = h/\sqrt{8mk_bT}$ es la longitud termica de de Broglie, $L_l$ es la longitud longitudinal de confinamiento, $L_*$ es una longitud caracteristica del dominio de confinamiento transversal y $\epsilon_{2D}$ son las autoenergias de la solución de la ecuación de Schrodinger en dos dimensiones para el confinamiento transversal. Los parametros usados en el articulo son: $T=300$ $K$, la longitud de cuadrado exterior e interior son de longitud  $21$ $nm$ y $13$ $nm$ respectivamente. Finalmente $\alpha_l=1$, $\alpha_t=0.005$. Por tanto, con estas consideraciones es posible hacer el calculo de tres propiedades termodinamicas relevantes para el estudio de los sistemas:

# Energia de libre de Helmholtz.

$$F = -k_b T  \text{ln} \left(\zeta^N/N! \right) = -k_b T  \text{ln} \left(\zeta^N/N! \right) =
 -k_b T N \text{ln} \left(\zeta\right)+k_b T \text{ln} \left(N!\right) $$

Usando una aproximacion muy conocida en física estadistica $\text{ln} \left(N!\right) \approx N \text{ln} \left(N\right)-N$. Computacionalmente graficamos,

$$\tilde{F} = F/Nk_bT = 
 -\text{ln} \left(\zeta\right)+ \text{ln} \left(N\right) - 1$$

De lo cual ignoraremos el termino $\text{ln} \left(N\right) - 1$ ya que solo contribuye en un reescalamiento del grafico.

Computacionalmente graficamos,

$$\tilde{S} = S/Nk_b =  \text{ln} \left(\zeta \right) + \left \{ \frac{1

# Solución ecuación de Schrodinger 

Como motivación, primero abordamos el problema de la solución de la ecuación de Schrodinger en 1D, ya que la extensión a 2D es similar. Para resolver numéricamente la ecuación de Schrodinger se emplea el método de diferencias finitas usando la aproximación de la segunda derivada conocida como diferencia central. En una dimensión, la ecuación de Schrodinger se expresa

\begin{equation}\tag{1}
E \psi(x) = -\frac{\hbar^2}{2m} \frac{\partial^{2} \psi(x)}{\partial x^{2}} + V(x)\psi(x),
\end{equation}

donde $\psi(x)$ es la función  de onda, V(x) el potencial y E la autoenergia del sistema. Computacionalmente solucionamos esta ecuación para un conjunto de N puntos. Por tanto, por simplicidad denotamos  $\psi(x) \rightarrow \psi_i$ con $i=1,2,...N $. La aproximación de diferencia central permite aproximar $\frac{\partial^{2} \psi(x)}{\partial x^{2}}$ como: 

\begin{equation}\tag{2}
\frac{\partial^{2} \psi(x)}{\partial x^{2}} = \frac{\psi_{i+1} - \psi_i + \psi_{i-1}}{\Delta x^{2}}.
\end{equation}

La ecuación de Schrodinger puede ser reescrita como

\begin{equation}\tag{1}
E \psi_i = -\frac{\hbar^2}{2m} (\frac{\psi_{i+1} - 2\psi_i + \psi_{i-1}}{\Delta x^{2}}) + V_i \psi_i
\end{equation}

Ahora, la estrategia es convertir la ecuación anterior en forma matricial. Sea $\bar{\psi}$ el vector columna que representa cada $\psi_i$,

\begin{equation}
\bar{\psi} = 
    \begin{pmatrix}
    \psi_1 \\
    \psi_2 \\
    \vdots \\
    \psi_i \\
    \vdots \\
    \psi_N \\
    \end{pmatrix}
\end{equation}

Podemos considerar cada valor de $i$ en el lado derecho de la ecuación como la matriz $\bar{J}$ tal que

\begin{equation}\tag{1}
\bar{J}\bar{\psi} = E\bar{\psi}
\end{equation}

De este modo sabemos que la matriz $\bar{J}$ debe tener dimensiones NxN. Asi por simple comparación, concluimos que,

\begin{equation}
\left[\frac{\hbar^{2}}{2m \Delta x^2}
    \begin{pmatrix}
    -2 & 1 & 0 & \cdots & 0\\
    1 & -2 & 1 & \cdots & \vdots\\
    0 & 1 & -2 & \cdots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    0 & \cdots & \cdots & \cdots & -2\\
    \end{pmatrix}
    + 
    \begin{pmatrix}
    V_1 & 0 & 0 & \cdots & 0\\
    0 & V_2 & 0 & \cdots & \vdots\\
    0 & 0 & V_3 & \cdots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    0 & \cdots & \cdots & \cdots & V_{N-1}\\
    \end{pmatrix}
\right]
\begin{pmatrix}
    \psi_1 \\
    \psi_2 \\
    \vdots \\
    \psi_i \\
    \vdots \\
    \psi_N \\
\end{pmatrix}
= E \begin{pmatrix}
    \psi_1 \\
    \psi_2 \\
    \vdots \\
    \psi_i \\
    \vdots \\
    \psi_N \\
\end{pmatrix}
\end{equation}

Por tanto, hallar los autovalores $E$, computacionalmente se reduce a diagonalizar la matriz definida entre corchetes. 

En el caso de dos dimensiones, se procede de forma analoga. La ecuación de Schrodinger puede ser escrita como 

\begin{equation}\tag{1}
E \psi(x, y) = \left[-\frac{\hbar^2}{2m} \nabla^2 + V(x, y)\right]\psi(x, y).
\end{equation}

Similar a como se realiza para una dimensión, se discretiza la función de onda en una malla rectangular, $\psi(x, y) \rightarrow \psi_{i,j}$ con $i,j =1,2...N$ y por tanto la ecuación de Schrodinger usando diferencia central,


\begin{equation}\tag{1}
E \psi_{i,j} = -\frac{\hbar^2}{2m} \left( \frac{\psi_{i+1,j} - 2\psi_{i,j} + \psi_{i-1,j}}{\Delta x^{2}} + \frac{\psi_{i,j+1} - 2\psi_{i,j} + \psi_{i,j-1}}{\Delta y^{2}} \right) + V_i \psi_{i,j}
\end{equation}

Para usar la misma metodología del caso 1d, buscamos escribir la ecuación anterior en forma $\bar{J}\bar{\psi} = E\bar{\psi}$. Para ello, ya que la función de onda es una matriz $N  \times N$, es necesario convertirla en un vector de tamaño $N^2 \times 1$ para eso, se ponen las filas de la matriz una debajo de otra comenzando con la primera. 

\begin{equation}

\begin{pmatrix}
    \psi_{11} & \psi_{12} & \cdots & \psi_{1N}\\
    \psi_{21} & \psi_{22} & \cdots & \psi_{2N}\\
    \vdots & \vdots & \ddots & \vdots\\
    \psi_{N1} & \psi_{N2} & \cdots & \psi_{NN}\\
    \end{pmatrix}
\rightarrow
    \begin{pmatrix}
        \psi_{11} \\
        \vdots \\
        \psi_{1N} \\
        \psi_{12} \\
        \vdots \\
        \psi_{2N} \\
        \vdots \\
        \psi_{NN} \\
    \end{pmatrix}
\end{equation}


Por otra parte, la matriz bar{J} debe de ser de dimensión $N^2 \times N^2$, y por analogia al caso 1D, concluimos que debe de ser una matriz diagonal por bloques, donde en cada bloque $N \times N$ esta compuesto por el operador $J$ una dimensión. Esto es,

$$\bar{J}= -\frac{\hbar^2}{2m} \left(D_{xx} + D_{yy} \right) + \bar{V}$$


Donde, el término $D_{xx}$  puede ser escrito como 

\begin{equation}
D_{xx} = 
\frac{1}{\Delta x^2}
    \begin{pmatrix}
    -2I & I & 0 & \cdots & 0\\
    I & -2I & I & \cdots & \vdots\\
    0 & I & -2I & \cdots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    0 & \cdots & \cdots & \cdots & -2I\\
    \end{pmatrix}
\end{equation}

Con I la matriz identidad $N \times N$. De manera similar,

\begin{equation}
D_{yy} = 
\frac{1}{\Delta y^2}
    \begin{pmatrix}
    -2I & I & 0 & \cdots & 0\\
    I & -2I & I & \cdots & \vdots\\
    0 & I & -2I & \cdots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    0 & \cdots & \cdots & \cdots & -2I\\
    \end{pmatrix}
\end{equation}


Así mismo, la energía potencial es también una matriz y puede ser escrita como 

\begin{equation}
\bar{V}  = 
\frac{1}{\Delta x^2}
    \begin{pmatrix}
    V_{11} & 0 & 0 & \cdots & 0\\
    0 & V_{12} & 0 & \cdots & \vdots\\
    0 & 0 & V_{13} & \cdots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    \vdots & \vdots & \vdots & \ddots & \vdots\\
    0 & \cdots & \cdots & \cdots & V_{NN}\\
    \end{pmatrix}
\end{equation}

Finalmente, con estas expresiones al diagonalizar $\bar{J}$ podemos obtener las autoenergias.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
from scipy.sparse import linalg as sla
from scipy import ndimage
import math

#constantes 
T = 300  #k 
densidad_particulas = 5e24 #m⁻3
Ll = 763
Lo = 21
Li = 13
L_ = 3.815
 
```

```python
def schrodinger2D(xmin, xmax, Nx, ymin, ymax, Ny,
                  Vfun2D, neigs, E0=0.0, findpsi=False):
    x = np.linspace(xmin, xmax, Nx)  
    dx = x[1] - x[0]  
    y = np.linspace(ymin, ymax, Ny)
    dy = y[1] - y[0]
 
    V = Vfun2D
 
    # create the 2D Hamiltonian matrix
    Hx = sparse.lil_matrix(2 * np.eye(Nx))
    for i in range(Nx - 1):
        Hx[i, i + 1] = -1
        Hx[i + 1, i] = -1
    Hx = Hx / (dx ** 2)
     
    Hy = sparse.lil_matrix(2 * np.eye(Ny))
    for i in range(Ny - 1):
        Hy[i, i + 1] = -1
        Hy[i + 1, i] = -1
    Hy = Hy / (dy ** 2)
 
    Ix = sparse.lil_matrix(np.eye(Nx))
    Iy = sparse.lil_matrix(np.eye(Ny))
    H = sparse.kron(Iy, Hx) + sparse.kron(Hy, Ix)  
 
    # Convert to lil form and add potential energy function
    H = H.tolil()
    for i in range(Nx * Ny):
        H[i, i] = H[i, i] + V[i]    
 
    # convert to csc form and solve the eigenvalue problem
    H = H.tocsc()  
    [evl, evt] = sla.eigs(H, k=neigs, sigma=E0)
             
    if findpsi == False:
        return evl
    else: 
        return evl, evt
```

```python
def particion(energia,  L_ = 3.815):  #Metodo 1 para calcular la función partición, suma sobre todas las energía posibles incluyendo las repetidas21
    energias=np.unique(energia, return_counts=True)
    alphat = 1
    alphal = 0.005
    h = 1
    m = 1 
    epsilon = 0
    for i in range(len(energias[0])):
        k_ = (2*m*energias[0][i]/h**2) * (L_**2)
        epsilon += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) 
    Z = epsilon * ((np.pi**(1/2))/(2*alphal) - 1/2) 
    return Z
```

```python
def n(ee,ev,TE,NN): 

    nn=0
    
    for i in range(len(ee)):
        evn=ev.T[i].reshape((NN,NN))
        nn += np.exp(-ee[i]/TE)*np.abs(evn)**2
    
    nn = nn/particion(ee,  L_ = 3.815)
    return nn
```

```python
def entropy(energia,  L_ = 3.8):
    energias=np.unique(energia, return_counts=True)
    alphat = 1
    alphal = 0.005
    h = 1
    m = 1 
    denom = 0
    numer = 0
    for i in range(len(energias[0])):

        k_ = (2*m*energias[0][i]/h**2) * (L_**2)

        denom += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) 
        
        numer += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) * k_
    
    term1 = np.sqrt(np.pi)/(4*alphal)*1/((np.pi**(1/2))/(2*alphal) - 1/2)
    term2 =  numer/denom * ((alphat**2)/(np.pi)**2) 

    return (term1+term2)+np.log(particion(energia,  L_ = 3.815))




```

```python
def twoD_to_oneD(Nx, Ny, F):
    # from a 2D matrix F return a 1D vector V
    V = np.zeros(Nx * Ny)
    vindex = 0
    for i in range(Ny):
        for j in range(Nx):
            V[vindex] = F[i, j]
            vindex = vindex + 1                     
    return V
```

```python
def square2D(x,y):
    Li=13
    n=len(x)
    m=len(y)
    R1=np.zeros([1,n])
    R2=np.zeros([1,n])
    R1[0,x>-Li/2]=1
    R2[0,x<Li/2]=1
    Rx=R1*R2
    R1=np.zeros([1,m])
    R2=np.zeros([1,m])
    R1[0,y>-Li/2]=1
    R2[0,y<Li/2]=1
    Ry=R1*R2
    return Rx.T@Ry



def rect2D(x,y):
    
    Li=13
    n=len(x)
    m=len(y)
    R1=np.zeros([1,n])
    R2=np.zeros([1,n])
    R1[0,x>-Li/8]=1
    R2[0,x<Li/8]=1
    Rx=R1*R2
    R1=np.zeros([1,m])
    R2=np.zeros([1,m])
    R1[0,y>-Li/8]=1
    R2[0,y<Li/8]=1
    Ry=R1*R2
    return Rx.T@Ry

```

# Gross - Pitaevskii equation.


```python
def schrodinger1D(xmin, xmax, Nx,
                  Vfun1D, neigs, E0=0.0, findpsi=False):
    x = np.linspace(xmin, xmax, Nx)  
    dx = x[1] - x[0]  
 
    V = Vfun1D
 
    # create the 2D Hamiltonian matrix
    Hx = sparse.lil_matrix(2 * np.eye(Nx))
    for i in range(Nx - 1):
        Hx[i, i + 1] = -1
        Hx[i + 1, i] = -1
    H = Hx / (2*dx ** 2)
    # Convert to lil form and add potential energy function
    H = H.tolil()
    for i in range(Nx):
        H[i, i] = H[i, i] + V[i]    
 
    # convert to csc form and solve the eigenvalue problem
    H = H.tocsc()  
    [evl, evt] = sla.eigs(H, k=neigs, sigma=E0)
             
    if findpsi == False:
        return evl
    else: 
        return evl, evt , x
```
