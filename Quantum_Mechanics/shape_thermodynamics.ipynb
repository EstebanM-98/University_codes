## Introduction

It is widely recognized that the thermodynamic properties of particle systems are intrinsically linked to the size of the domain containing them. For example, the internal energy of particles in a closed rectangular box experiences variations as the size of the box, that is, its volume, is adjusted. In the article titled <i>"Quantum shape effects and novel thermodynamic behaviors at the nanoscale"</i>, researchers Alhun Aydin and Altug Sisman from Uppsala University in Sweden highlight that the geometric configuration of the domains that house the Particles play a crucial role in determining thermodynamic state variables. In other words, these parameters are conditioned not only by the size of the domains but also by their shape. It is evident that by altering the shape of the rectangular box without modifying its volume, the thermodynamic properties of the gas in question undergo significant changes. This study provides a novel perspective by exploring how confinement geometry influences nanometer-scale thermodynamic behavior.

The study's authors devise specific confinement domains to illustrate that, by modifying the shape of these domains without altering their size, thermodynamic potentials, such as Helmholtz free energy, entropy, and internal energy, undergo significant variations. This approach demonstrates the direct influence of the geometric configuration on the system's thermodynamic properties, highlighting the importance of considering the shape of the domain as a crucial factor in determining its thermodynamic behavior. This finding triggers new perspectives in understanding, offering valuable contributions to the field of thermodynamics in reduced dimensions.

This notebook aims to reproduce the article's results, focusing primarily on figures 2 and 3. First, the theory necessary to solve the problem is built, starting with calculating the thermodynamic potentials, passing through the partition function, and finally defining the method. To numerically solve the Schrodinger equation since energy eigenstates are vital for our purpose. ## Methods.

1. Calculation of thermodynamic properties.

To visualize the variations in the thermodynamic potentials with the changes in the shape of the confinement domain, two potentials that confine the gas under study have been created. In the first figure, an internal cube has been designed to be housed inside an external cube of larger dimensions. The gas occupies the space between both cubes, located inside the external cube but outside the internal cube. The second figure presents an internal parallelepiped contained within another of higher dimensions, with the gas between the internal and external parallelepiped. It is essential to highlight the specific lengths of each figure to understand its configuration. Changes in the shape of the domain are implemented by rotating the internal figures by an angle $\theta$, which modifies the space structure that houses the particles without altering their size. This experimental approach allows us to explore in detail how the domain geometry affects the thermodynamic properties of the system without changing its volume.

To calculate the thermodynamic potentials with these two geometries, the partition function of the system must be found. In statistical physics, for a system in equilibrium consisting of N particles, there is an object that relates the microscopic quantities of the system with its thermodynamic properties. This object is known as the partition function, which is given by:

$$Z = \sum_\epsilon \text{exp}\left(-\beta \epsilon\right),$$

where $\beta=1/k_bT$ and $\epsilon$ are the self-energies of the system. Our study system is an ideal gas confined in the two geometries described above, shown below:

Considering that the system is non-interacting, the partition function can be calculated more simply and simpler since the system's total energy is separable from the kinetic energy contribution of each particle. Therefore:

$$Z = \left \{ \sum_\epsilon \text{exp}\left(-\beta \epsilon\right) \right\}^N /N!,$$

The factor $N!$ is added when considering the indistinguishability of the particles. Since the particles are confined in 3D, the Schrodinger equation should be solved for three dimensions. However, in the article, an approximation is used referring to the Weyl hypothesis, where the self-energies are approximated in the longitudinal direction, arriving at,

$$\zeta=\left(\frac{\sqrt{\pi}}{2 \alpha_l}-\frac{1}{2}\right)\sum_{k}\text{exp}\left(-\ frac{\alpha _{t}^2}{\pi^2} \tilde{k}^2 \right),$$

and the total partition function $ Z= \zeta^{N}/N!$. The parameters $\alpha_t = L_c/L_*$, $\alpha_l=L_c/L_l$ and $\tilde{k}= \frac{2m\epsilon_{2D}}{\hbar^2}$ are defined. Here $L_c = h/\sqrt{8mk_bT}$ is the thermal de Broglie length, $L_l$ is the longitudinal confinement length, $L_*$ is a characteristic length of the transverse confinement domain, and $\epsilon_{2D} $ are the self-energies of the solution of the Schrodinger equation in two dimensions for transverse confinement. The parameters used in the article are $T=300$ $K$, the outer and inner square length $21$ $nm$, and $13$ $nm$, respectively. Finally $\alpha_l=1$, $\alpha_t=0.005$. Therefore, with these considerations, it is possible to calculate three thermodynamic properties relevant to the study of systems:

## 1.1 Helmholtz free energy.

$$F = -k_b T \text{ln} \left(\zeta^N/N! \right) = -k_b T \text{ln} \left(\zeta^N/N! \right) =
  -k_b T N \text{ln} \left(\zeta\right)+k_b T \text{ln} \left(N!\right) $$

Using the approximation $\text{ln} \left(N!\right) \approx N \text{ln} \left(N\right)-N$. Computationally, we graph,

$$\tilde{F} = F/Nk_bT =
  -\text{ln} \left(\zeta\right)+ \text{ln} \left(N\right) - 1$$

We will ignore the term $\text{ln} \left(N\right) - 1$ since it only contributes to rescaling the graph.

## 2. Solution of Schrodinger's equation

## 2.1 Finite difference method

To calculate the system's potentials and, therefore, the partition function, we saw that it is necessary to know the energy eigenstates of the system; for this, we solved the Schrodinger equation in two dimensions. As motivation, we first address the problem of solving the Schrodinger equation in 1D since the extension to 2D is similar. To numerically solve the Schrodinger equation, the finite difference method uses the second derivative approximation, the central difference. In one dimension, the Schrodinger equation is expressed.

\begin{equation}\tag{1}
E \psi(x) = -\frac{\hbar^2}{2m} \frac{\partial^{2} \psi(x)}{\partial x^{2}} + V(x)\psi (x),
\end{equation}

where $\psi(x)$ is the wave function, V(x) the potential and E the self-energy of the system. Computationally, we solve this equation for a set of N points. Therefore, for simplicity we denote $\psi(x) \rightarrow \psi_i$ with $i=1,2,...N $. The central difference approximation allows us to approximate $\frac{\partial^{2} \psi(x)}{\partial x^{2}}$ as:

\begin{equation}\tag{2}
\frac{\partial^{2} \psi(x)}{\partial x^{2}} = \frac{\psi _{i+1} - \psi_i + \psi _{i-1}}{\Delta x^{2}}.
\end{equation}

Schrodinger's equation can be rewritten as

\begin{equation}\tag{1}
E \psi_i = -\frac{\hbar^2}{2m} (\frac{\psi _{i+1} - 2\psi_i + \psi _{i-1}}{\Delta x^{2}}) + V_i \psi_i
\end{equation}

Now, the strategy is to convert the above equation into matrix form. Let $\bar{\psi}$ be the column vector that represents each $\psi_i$,

\begin{equation}
\bar{\psi} =
     \begin{pmatrix}
     \psi_1 \\
     \psi_2 \\
     \vdots \\
     \psi_i \\
     \vdots \\
     \psi_N \\
     \end{pmatrix}
\end{equation}

We can consider each value of $i$ on the right side of the equation as the matrix $\bar{J}$ such that.

\begin{equation}\tag{1}
\bar{J}\bar{\psi} = E\bar{\psi}
\end{equation}

In this way, we know that the matrix $\bar{J}$ must have dimensions NxN. Thus, by simple comparison, we conclude that,

\begin{equation}
\left[\frac{\hbar^{2}}{2m \Delta x^2}
     \begin{pmatrix}
     -2 & 1 & 0 & \cdots & 0\\
     1 & -2 & 1 & \cdots & \vdots\\
     0 & 1 & -2 & \cdots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     0 & \cdots & \cdots & \cdots & -2\\
     \end{pmatrix}
     +
     \begin{pmatrix}
     V_1 & 0 & 0 & \cdots & 0\\
     0 & V_2 & 0 & \cdots & \vdots\\
     0 & 0 & V_3 & \cdots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     0 & \cdots & \cdots & \cdots & V_{N-1}\\
     \end{pmatrix}
\right]
\begin{pmatrix}
     \psi_1 \\
     \psi_2 \\
     \vdots \\
     \psi_i \\
     \vdots \\
     \psi_N \\
\end{pmatrix}
= E \begin{pmatrix}
     \psi_1 \\
     \psi_2 \\
     \vdots \\
     \psi_i \\
     \vdots \\
     \psi_N \\
\end{pmatrix}
\end{equation}

Therefore, finding the eigenvalues $E$ computationally is reduced to diagonalizing the matrix between brackets.

In the case of two dimensions, the procedure is analogous. Schrodinger's equation can be written as

\begin{equation}\tag{1}
E \psi(x, y) = \left[-\frac{\hbar^2}{2m} \nabla^2 + V(x, y)\right]\psi(x, y).
\end{equation}

Like doing it for one dimension, the wave function is discretized in a rectangular mesh, $\psi(x, y) \rightarrow \psi _{i,j}$ with $i,j =1,2... N$ and therefore the Schrodinger equation using central difference,


\begin{equation}\tag{1}
E \psi _{i,j} = -\frac{\hbar^2}{2m} \left( \frac{\psi _{i+1,j} - 2\psi _{i,j} + \psi _{i -1,j}}{\Delta x^{2}} + \frac{\psi _{i,j+1} - 2\psi _{i,j} + \psi _{i,j-1}}{\ Delta y^{2}} \right) + V_i \psi _{i,j}
\end{equation}

To use the same methodology as case 1d, we seek to write the previous equation in the form $\bar{J}\bar{\psi} = E\bar{\psi}$. Since the wave function is a matrix $N \times N$, it is necessary to convert it into a vector of size $N^2 \times 1$. For this, the matrix rows are placed one below the other, starting with the first.

\begin{equation}

\begin{pmatrix}
     \psi_{11} & \psi_{12} & \cdots & \psi_{1N}\\
     \psi_{21} & \psi_{22} & \cdots & \psi_{2N}\\
     \vdots & \vdots & \ddots & \vdots\\
     \psi_{N1} & \psi_{N2} & \cdots & \psi_{NN}\\
     \end{pmatrix}
\rightarrow
     \begin{pmatrix}
         \psi_{11} \\
         \vdots \\
         \psi_{1N} \\
         \psi_{12} \\
         \vdots \\
         \psi_{2N} \\
         \vdots \\
         \psi_{NN} \\
     \end{pmatrix}
\end{equation}


On the other hand, the matrix bar{J} must be of dimension $N^2 \times N^2$. By analogy to the 1D case, we conclude that it must be a diagonal matrix by blocks, wherein each block $N \times N$ is composed of the one-dimensional operator $J$. This is,

$$\bar{J}= -\frac{\hbar^2}{2m} \left(D_{xx} + D_{yy} \right) + \bar{V}$$


Where, the term $D_{xx}$ can be written as

\begin{equation}
D_{xx} =
\frac{1}{\Delta x^2}
     \begin{pmatrix}
     -2I & I & 0 & \cdots & 0\\
     I & -2I & I & \cdots & \vdots\\
     0 & I & -2I & \cdots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     0 & \cdots & \cdots & \cdots & -2I\\
     \end{pmatrix}
\end{equation}

With I the identity matrix $N \times N$. Similarly,

\begin{equation}
D_{yy} =
\frac{1}{\Delta y^2}
     \begin{pmatrix}
     -2I & I & 0 & \cdots & 0\\
     I & -2I & I & \cdots & \vdots\\
     0 & I & -2I & \cdots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     0 & \cdots & \cdots & \cdots & -2I\\
     \end{pmatrix}
\end{equation}


Likewise, potential energy is also a matrix and can be written as

\begin{equation}
\bar{V} =
\frac{1}{\Delta x^2}
     \begin{pmatrix}
     V_{11} & 0 & 0 & \cdots & 0\\
     0 & V_{12} & 0 & \cdots & \vdots\\
     0 & 0 & V_{13} & \cdots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     \vdots & \vdots & \vdots & \ddots & \vdots\\
     0 & \cdots & \cdots & \cdots & V_{NN}\\
     \end{pmatrix}
\end{equation}

Finally, with these expressions by diagonalizing $\bar{J}$ we can obtain the self-energies.

```python
#importacion de librerías requeridas
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
from scipy.sparse import linalg as sla
from scipy import ndimage
import math

#constantes 
T = 300  #k 
densidad_particulas = 5e24 #m⁻3
Ll = 763  #nm
Lo = 21  #nm
Li = 13  #nm
L_ = 3.815  #nm
```

```python
#implementación del método de diferencias finitas 
def schrodinger2D(xmin, xmax, Nx, ymin, ymax, Ny,
                  Vfun2D, neigs, E0=0.0, findpsi=False):
    x = np.linspace(xmin, xmax, Nx)  
    dx = x[1] - x[0]  
    y = np.linspace(ymin, ymax, Ny)
    dy = y[1] - y[0]
 
    V = Vfun2D
 
    # create the 2D Hamiltonian matrix
    Hx = sparse.lil_matrix(2 * np.eye(Nx))
    for i in range(Nx - 1):
        Hx[i, i + 1] = -1
        Hx[i + 1, i] = -1
    Hx = Hx / (dx ** 2)
     
    Hy = sparse.lil_matrix(2 * np.eye(Ny))
    for i in range(Ny - 1):
        Hy[i, i + 1] = -1
        Hy[i + 1, i] = -1
    Hy = Hy / (dy ** 2)
 
    Ix = sparse.lil_matrix(np.eye(Nx))
    Iy = sparse.lil_matrix(np.eye(Ny))
    H = sparse.kron(Iy, Hx) + sparse.kron(Hy, Ix)  
 
    H = H.tolil()
    for i in range(Nx * Ny):
        H[i, i] = H[i, i] + V[i]    
 
    H = H.tocsc()  
    [evl, evt] = sla.eigs(H, k=neigs, sigma=E0)
             
    if findpsi == False:
        return evl
    else: 
        return evl, evt
```

```python
#Metodo para calcular la función partición, suma sobre todas las energía posibles incluyendo las repetidas, tiene en cuenta la degeneración del espectro
def particion(energia,  L_ = 3.815): 
    energias=np.unique(energia, return_counts=True)
    alphat = 1
    alphal = 0.005
    h = 1 #constante de plack igualada a 1
    m = 1 #masa de la particula igualada a 1
    epsilon = 0
    for i in range(len(energias[0])):
        k_ = (2*m*energias[0][i]/h**2) * (L_**2)
        epsilon += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) 
    Z = epsilon * ((np.pi**(1/2))/(2*alphal) - 1/2) 
    return Z
```

```python
#Para visualizar la densidad de partículas
def n(ee,ev,TE,NN): 
    nn=0
    for i in range(len(ee)):
        evn=ev.T[i].reshape((NN,NN))
        nn += np.exp(-ee[i]/TE)*np.abs(evn)**2
    
    nn = nn/particion(ee,  L_ = 3.815)
    return nn
```

```python
#función para calcular la entropía
def entropy(energia,  L_ = 3.8):
    energias=np.unique(energia, return_counts=True)
    alphat = 1
    alphal = 0.005
    h = 1
    m = 1 
    denom = 0
    numer = 0
    for i in range(len(energias[0])):

        k_ = (2*m*energias[0][i]/h**2) * (L_**2)

        denom += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) 
        
        numer += energias[1][i] * np.exp(-((alphat**2)/(np.pi)**2) * k_) * k_
    
    term1 = np.sqrt(np.pi)/(4*alphal)*1/((np.pi**(1/2))/(2*alphal) - 1/2)
    term2 =  numer/denom * ((alphat**2)/(np.pi)**2) 

    return (term1+term2)+np.log(particion(energia,  L_ = 3.815))
```

```python
# función para pasar de matriz 2D a vector 1D
def twoD_to_oneD(Nx, Ny, F):
    V = np.zeros(Nx * Ny)
    vindex = 0
    for i in range(Ny):
        for j in range(Nx):
            V[vindex] = F[i, j]
            vindex = vindex + 1                     
    return V
```

```python
#cubo
def square2D(x,y):
    Li=13
    n=len(x)
    m=len(y)
    R1=np.zeros([1,n])
    R2=np.zeros([1,n])
    R1[0,x>-Li/2]=1
    R2[0,x<Li/2]=1
    Rx=R1*R2
    R1=np.zeros([1,m])
    R2=np.zeros([1,m])
    R1[0,y>-Li/2]=1
    R2[0,y<Li/2]=1
    Ry=R1*R2
    return Rx.T@Ry

#paralelepípedo
def rect2D(x,y):
    
    Li=13
    n=len(x)
    m=len(y)
    R1=np.zeros([1,n])
    R2=np.zeros([1,n])
    R1[0,x>-Li/8]=1
    R2[0,x<Li/8]=1
    Rx=R1*R2
    R1=np.zeros([1,m])
    R2=np.zeros([1,m])
    R1[0,y>-Li/8]=1
    R2[0,y<Li/8]=1
    Ry=R1*R2
    return Rx.T@Ry

```

```python
angle=np.linspace(0,90,90)
F = []
S = []
U = []
neigs = 35

for i in range(len(angle)):
    
    rotated_img = ndimage.rotate(square,angle[i],reshape=False)
    Vfun2D = twoD_to_oneD(Nx, Ny, rotated_img )
    EE = schrodinger2D(min(x1), max(x1), Nx, min(y1), max(y1), Ny,
                  Vfun2D, neigs, findpsi=False) 
    Z = particion(EE)
    s = entropy(EE,  L_ = 3.815)
    f = -np.log(Z)
    F.append(f)
    S.append(s)
    U.append(f+s)


ang = [0,10,25,45]
ns = []
for i in range(0,4):
    
    rotated_img = ndimage.rotate(square,ang[i],reshape=False)
    Vfun2D = twoD_to_oneD(Nx, Ny, rotated_img)
    EE , EV = schrodinger2D(min(x1), max(x1), Nx, min(y1), max(y1), Ny,
                  Vfun2D, neigs, findpsi=True) 
    density=n(EE, EV, 300, Nx)
    ns.append(density)
    
```
