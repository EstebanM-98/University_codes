```python
# LIBRERIAS
import matplotlib
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy.interpolate import CubicSpline
#matplotlib.rc('xtick', labelsize=20) 
#matplotlib.rc('ytick', labelsize=20) 
font = {'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)

```

```python
c=3*1e8 #Velocidad de luz
landa_c=1550*1e-9  #longitud de onda
v=0.17   #Constante
k_onda=2*np.pi/landa_c  #k de onda careverga
DeltaP=0.15 #Constante del articulo
k=0.5*1e2 #Unidades cm^{-1}
n = 1.46  #Indice de refracción
d1=125*1e-6 #Diametro exterioir um
r=d1/2
Beta = 0.025*0.25*(k_onda)*(n**3)*(DeltaP)*(1 + v)*(k**2)*(r**2)
D=56*1e-3#a=7.7*1e7*(k**2)*d2**2
landa=1e-9*np.linspace(300,2100,5000)
N=np.array([1,2,3,4,5,6])
d2=125e-6
#New beta
Beta2=7.7*1e7*(k**2)*r**2
```

 # Configuración thorlabs, figura 3.
 
 Plot of the retardance per paddle for silica fiber with Ø125 µm cladding on the FPC560, which has a loop diameter of 56 mm
 
 -Esteban, el diametro de nuestras paletas es de 38 mm

```python
#calibración thorlabs


def desfase(landa,N,Beta):
    return ((np.pi)*N*Beta*d2**2)/(landa*D)

plt.plot(landa*1e9,desfase(landa,N[0],Beta),label="1 loop")
plt.plot(landa*1e9,desfase(landa,N[1],Beta),label="2 loop")
plt.plot(landa*1e9,desfase(landa,N[2],Beta),label="3 loop")
plt.plot(landa*1e9,desfase(landa,N[3],Beta),label="4 loop")
plt.plot(landa*1e9,desfase(landa,N[4],Beta),label="5 loop")
plt.plot(landa*1e9,desfase(landa,N[5],Beta),label="6 loop")
plt.xlim(300,2100)

plt.grid()
plt.legend()
plt.title("Desfase")
plt.xlabel("wavelength(nm)")
plt.ylabel("Retardance(Waves)")
```

```python
plt.plot(landa*1e9,desfase(landa,N[0],Beta2),label="1 loop")
plt.plot(landa*1e9,desfase(landa,N[1],Beta2),label="2 loop")
plt.plot(landa*1e9,desfase(landa,N[2],Beta2),label="3 loop")
plt.plot(landa*1e9,desfase(landa,N[3],Beta2),label="4 loop")
plt.plot(landa*1e9,desfase(landa,N[4],Beta2),label="5 loop")
plt.plot(landa*1e9,desfase(landa,N[5],Beta2),label="6 loop")
plt.xlim(300,2100)

plt.grid()
plt.legend()
plt.title("Desfase beta2")
plt.xlabel("wavelength(nm)")
plt.ylabel("Retardance(Waves)")
```
